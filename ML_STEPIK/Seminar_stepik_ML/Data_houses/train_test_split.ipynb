# Train-Test

```python
import pandas as pd
import sys
import random
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Advertising.csv')
```

```python
df.head()
```

```python
X = df.drop('sales', axis = 1)
y = df['sales']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler_train = StandardScaler()
scaler_train.fit(X_train)
```

```python
X_train = scaler_train.transform(X_train)
```

```python
scaler_test = StandardScaler()
scaler_test.fit(X_test)
```

```python
X_test = scaler_test.transform(X_test)
```

```python
from sklearn.linear_model import Ridge
```

```python
model = Ridge(alpha = 100)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test, y_pred)
```

```python
model_2 = Ridge(alpha = 1)
model_2.fit(X_train, y_train)
y_pred = model_2.predict(X_test)
mean_squared_error(y_test, y_pred)
```

# Train-Validation-Test

```python
import pandas as pd
import sys
import random
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Advertising.csv')
```

```python
df.head()
```

```python
X = df.drop('sales', axis = 1)
```

```python
y = df['sales']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_other, y_train, y_other = train_test_split(X, y, test_size = 0.5, random_state = 101)
```

```python
X_eval, X_test, y_eval, y_test = train_test_split(X_other, y_other, test_size = 0.3, random_state = 101)
```

```python
X_eval.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
```

```python
X_eval = scaler.transform(X_eval)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import Ridge
```

```python
model_1 = Ridge(alpha = 100)
model_1.fit(X_train, y_train)
y_eval_pred = model_1.predict(X_eval)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_eval, y_eval_pred)
```

```python
model_2 = Ridge(alpha = 1)
model_2.fit(X_train, y_train)
y_eval_pred = model_2.predict(X_eval)
mean_squared_error(y_eval, y_eval_pred)
```

```python
y_final_pred = model_2.predict(X_test)
mean_squared_error(y_test, y_final_pred)
```
